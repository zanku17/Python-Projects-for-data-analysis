```python
# Formating the time column by removing ":"

def format_time(x):
    if ":" in x.lower():
        if int(x.split(":")[0]) >= 12 and int(x.split(":")[0]) < 18:
            x = "Afternoon"
        elif int(x.split(":")[0]) < 12:
            x = "Morning"
        elif int(x.split(":")[0]) >= 18:
            x = "Night"
    elif "morning" in x.lower() or "dawn" in x.lower():
        x = "Morning"
    elif "afternoon" in x.lower():
        x = "Afternoon"
    elif "night" in x.lower():
        x = "Night"
    else:
        x = "Not Known"
    return x
```

```python
land.time = land.time.apply(format_time)
```

```python
land.time.value_counts()
```
